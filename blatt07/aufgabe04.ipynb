```python
from rdflib import Graph
g = Graph()
g.parse("http://bigasterisk.com/foaf.rdf")

def inspect(graph):
    print("Anzahl der Statements: ", len(graph))

    for s, p, o in graph:
        print("Subject: {0}, Predicate: {1}, Object: {2}".format(s, p, o))

def check_name(graph):
    query = """
    PREFIX foaf: <http://xmlns.com/foaf/0.1/>
    SELECT ?givenname

    WHERE
    {
        ?x foaf:givenname ?givenname.
    }
    """

    qres = graph.query(query)
    return len(qres) > 0

def extract(graph):
    # query = """
    # PREFIX foaf: <http://xmlns.com/foaf/0.1/>
    # SELECT ?name ?homepage

    # WHERE
    # {
    #     ?x foaf:name ?name.
    #     ?x foaf:homepage ?homepage.
    # }
    # """

    # qres = graph.query(query)
    # for row in qres:
    #     print(f"Name: {row.name}, Homepage: {row.homepage}")
    
    query_name = """
    PREFIX foaf: <http://xmlns.com/foaf/0.1/>
    SELECT ?name
    
    WHERE
    {
        ?x foaf:name ?name.
    }
    """

    query_homepage = """
    PREFIX foaf: <http://xmlns.com/foaf/0.1/>
    SELECT ?homepage
    
    WHERE
    {
        ?x foaf:homepage ?homepage.
    }
    """
    
    qres_name = graph.query(query_name)
    qres_homepage = graph.query(query_homepage)

    names = ", ".join(str(x.name) for x in qres_name)
    homepages = ", ".join(str(x.homepage) for x in qres_homepage)

    print("Names: ", names)
    print("Homepages: ", homepages)


print("inspect:")
inspect(g)
print("\ncheck_name:")
print(check_name(g))
print("\nextract:")
extract(g)
```
