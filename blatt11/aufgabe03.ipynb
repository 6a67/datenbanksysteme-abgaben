```python
import pysubgroup as ps
import pandas as pd


def find_subgroups(database, targetInput): 
    # read pandas dataframe from file
    data = pd.read_csv(database, delimiter=';')

    target = ps.NumericTarget(targetInput)
    searchspace = ps.create_selectors(data, ignore=['G1', 'G2', 'G3'])
    task = ps.SubgroupDiscoveryTask (
        data,
        target,
        searchspace,
        result_set_size=10,
        depth=2,
        qf=ps.StandardQFNumeric(0.5)
    )
    result = ps.BeamSearch().execute(task)
    return result.to_dataframe()


print(find_subgroups('student-mat.csv', 'G1'))
print(find_subgroups('student-mat.csv', 'G2'))
print(find_subgroups('student-por.csv', 'G1'))
print(find_subgroups('student-por.csv', 'G2'))
```
